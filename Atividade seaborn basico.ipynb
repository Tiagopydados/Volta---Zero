```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('b1abb33c4f.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
nomes = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for nom in nomes:
    plt.figure(figsize=(6, 4))
    sns.barplot(x=nom, y=df.index, data=df)
    plt.title(f'Gráfico de barras para {nom}')
    plt.xlabel(nom)
    plt.ylabel('Índice')
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(12,8))
sns.barplot(x='CONSULTAS',y='APGAR5',data=df)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2,2, figsize=(18,10))
plt.subplot(2,2,1)
sns.countplot(x='SEXO',data=df)
plt.subplot(2,2,2)
sns.countplot(x='GESTACAO',data=df)
plt.subplot(2,2,3)
sns.countplot(x='ESTCIVMAE',data=df)
plt.subplot(2,2,4)
sns.countplot(x='CONSULTAS',data=df)
```
